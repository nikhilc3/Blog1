- Create 3 quiz questions to demonstrate the main steps in creating a repository on Github (.1)

What is the first step in creating a new repository on GitHub?

1. Click on the "Create new file" button
2. Click on the "New repository" button
3. Go to the "Pull requests" tab
4. Go to the "Code" tab

Correct Answer: b) Click on the "New repository" button

When creating a new repository, what should you enter in the "Repository name" field?

1. Your username
2. A unique name for your repository
3. The name of an existing repository
4. The name of your favorite coding language

Correct Answer: b) A unique name for your repository

What does the "Initialize this repository with a README" option do when creating a new GitHub repository?

1. It automatically creates a new branch
2. It automatically generates a basic README file in the repository
3. It connects the repository to a remote server
4. It protects the repository with a password

Correct Answer: b) It automatically generates a basic README file in the repository

    async function fetchSportsData() {
        const url = 'https://sportscore1.p.rapidapi.com/sports/1/teams';
        const params = { page: '1' };
        const headers = {
            'X-RapidAPI-Key': '6cf0c105b2mshabcf75f48cea6f9p14945ajsn30b6ebcd6ead',
            'X-RapidAPI-Host': 'sportscore1.p.rapidapi.com'
        };
        
        try {
            const response = await fetch(url + '?' + new URLSearchParams(params), {
                headers: headers
            });
            const data = await response.json(); // Getting the data as JSON
            console.log(data);
            document.getElementById('tips').innerHTML = JSON.stringify(data, null, 2); // Pretty printing the JSON data
        } catch (error) {
            console.error(error);
        }
    }
    </script>
